```python
text="Books—they have the power to transport readers to different worlds, eras, and emotions! Whether it’s a thrilling mystery... or a heartwarming tale, every story leaves its mark. Reading not only entertains, but also expands vocabulary, sparks imagination, and inspires thought. In today’s screen-dominated world, books offer something rare: a quiet, mindful escape. The joy of flipping pages, feeling the texture, and getting lost in a narrative—it's simply unmatched by any other medium."
print(text)
```

```python
from nltk.probability import FreqDist
freq_dist=FreqDist(filtered_tokens)

for word, freq in freq_dist.most_common(10):
    print(f"{word}: {freq}")

freq_dist.plot(10, title="Word Frequency (Excluding Stopwords)")
```

```python
from sklearn.feature_extraction.text import CountVectorizer
texts = ["AI revolutionizes healthcare with early disease detection","Global markets tumble amid economic uncertainty","New space mission aims to explore distant planets"]
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(texts) 
print(vectorizer.get_feature_names_out())
print(bow.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
X=tfidf.fit_transform(texts)
print(tfidf.get_feature_names_out())
print(X.toarray())
```

```python
import numpy as np
feature_names=tfidf.get_feature_names_out()

for i, row in enumerate(X.toarray()):
    top_indices=np.argsort(row)[::-1][:3]
    print(f"Text {i+1} Top 3 Keywords:")
    for index in top_indices:
        print(f"   {feature_names[index]}: {row[index]:.3f}")
    print()
```

Q4. Write 2 short texts (4–6 lines each) describing two different technologies (e.g., AI vs 
Blockchain). 
1. Preprocess and tokenize both texts. 
2. Calculate:
   
   a. Jaccard Similarity using sets

   
   b. Cosine Similarity using TfidfVectorizer + cosine_similarity()

   
   c. Analyze which similarity metric gives better insights in your case.

```python
text1 = """
Artificial Intelligence (AI) is a branch of computer science focused on creating systems capable of performing tasks that typically require human intelligence. 
AI systems are designed to learn, adapt, and make decisions based on data. Common applications of AI include machine learning, natural language processing, 
and robotics.
"""

text2 = """
Blockchain technology is a decentralized digital ledger system that records transactions across many computers in such a way that the registered transactions 
cannot be altered retroactively. Blockchain is the backbone of cryptocurrencies, providing secure and transparent transaction validation.
"""
```

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    tokens = word_tokenize(text)
    return tokens
```

```python
text1_p=preprocess_text(text1)
text2_p=preprocess_text(text2)
```

```python
print(text1_p)
print(text2_p)
```

```python
t1 = set(text1_p)
t2 = set(text2_p)
jaccard = len(t1&t2) / len(t1|t2)
print("Jaccard Similarity:", jaccard)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
tfidf_vec = TfidfVectorizer()
vecs = tfidf_vec.fit_transform([text1,text2])
cos_sim = cosine_similarity(vecs[0:1], vecs[1:2])
print("Cosine Similarity:", cos_sim[0][0])
```

Q5. Write a short review for a product or service. 
1. Use TextBlob or VADER to find polarity & subjectivity for each review. 
2. Classify reviews as positive, negative, or neutral. 
3. Create a word cloud using the wordcloud library for all positive reviews. 

```python
%pip install textblob
%pip install vaderSentiment
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

reviews = [
    "This product is absolutely fantastic! Highly recommend it.",
    "It's okay, not the best but not the worst either.",
    "Terrible experience. Would never buy again.",
    "Loved it! Excellent quality and fast delivery.",
    "Meh, it was average at best.",
    "Superb! Totally worth the money."
]
```

```python
def classify(compound_score):
    if compound_score>=0.05:
        print("Classification: Positive")
    elif compound_score<=-0.05:
        print("Classification: Negative")
    else:
        print("Classification: Neutral")
```

```python
for review in reviews:
    blob = TextBlob(review)
    print("Polarity:", blob.sentiment.polarity)
    print("Subjectivity:", blob.sentiment.subjectivity)
    sentiment = analyzer.polarity_scores(review)
    print("Sentiment Scores:", sentiment)
    compound_score = sentiment['compound']
    classify(compound_score)
    print()
```

```python
positive=""

for review in reviews:
    sentiment=analyzer.polarity_scores(review)
    if(sentiment['compound']>=0.5):
        positive+=" "+review
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Positive Reviews")
plt.show()
```

Q6. Choose your own paragraph (~100 words) as training data. 
1. Tokenize text using Tokenizer() from keras.preprocessing.text 
2. Create input sequences and build a simple LSTM or Dense model 
3. Train the model and generate 2–3 new lines of text starting from any seed word you provide.

```python
text = """The sunrise painted the sky with brilliant shades of orange and pink. 
Birds chirped cheerfully as the town slowly woke up from its slumber. 
The air was crisp, filled with the scent of dew, fresh earth, and blooming flowers. 
A gentle breeze rustled the trees, and the world felt peaceful and calm. 
Children laughed as they skipped to school, their backpacks bouncing behind them. 
Shopkeepers opened wooden shutters, arranging colorful goods with hopeful smiles. 
The aroma of baked bread drifted from a nearby café, mingling with the morning air. 
Dogs barked playfully in the distance, chasing each other along cobbled streets. 
It was the kind of morning that made people pause, breathe deeply, and appreciate the beauty in simple moments."""
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1
```

```python
token_list = tokenizer.texts_to_sequences([text])[0]
input_sequences = []
for i in range(1, len(token_list)):
    n_gram_sequence = token_list[:i+1]
    input_sequences.append(n_gram_sequence)
```

```python
padded = pad_sequences(input_sequences, padding='pre')
X = padded[:, :-1]
y = padded[:, -1]
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
model = Sequential()
model.add(Embedding(input_dim=total_words, output_dim=10))
model.add(LSTM(50))
model.add(Dense(50, activation='relu'))
model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')
model.summary()
```

```python
import numpy as np

def generate_text(seed_text, next_words=15):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=X.shape[1], padding='pre')
        predicted = model.predict(token_list, verbose=0)
        predicted_index = np.argmax(predicted)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted_index:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text

print("\nGenerated Text:\n")
print(generate_text("sunrise"))
```
